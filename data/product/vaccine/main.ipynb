```python
df = pd.DataFrame(vaccine_list, columns=columns)

date_cols = ["entry_date", "expiry_date", "manufacture_date"]
for col in date_cols:
    df[col] = df[col].astype(str)

# Ensure numeric
df["quantity"] = (
    pd.to_numeric(df["quantity"], errors="coerce")
      .fillna(1)
      .astype(int)
      .clip(lower=1)
)
df["price"] = pd.to_numeric(df["price"], errors="coerce").fillna(0)

discrete_columns = [
    "vaccine_name",
    "entry_date",
    "expiry_date",
    "manufacture_date"
]

model = CTGAN(epochs=500)

model.fit(df, discrete_columns=discrete_columns)

synthetic_df = model.sample(10000)

for col in date_cols:
  synthetic_df[col] = pd.to_datetime(
      synthetic_df[col],
      errors="coerce"
  )

full_df = pd.concat([df, synthetic_df], ignore_index=True)
full_df["quantity"] = abs(full_df["quantity"].astype(int))

print(full_df.info())
print(len(full_df))
```

```python
full_df["quantity"] += 1

def sql_date(value):
    if value is None or pd.isna(value):
        return "NULL"

    # N·∫øu l√† string
    if isinstance(value, str):
        try:
            value = pd.to_datetime(value, errors="coerce")
        except:
            return "NULL"

    if pd.isna(value):
        return "NULL"

    return f"'{value.date()}'"
    
def sql_escape(value):
    if value is None:
        return "NULL"
    return "'" + str(value).replace("'", "''") + "'"

def write_sql(vaccine_list):
    filename = "insert_vaccine.sql"

    with open(filename, "w", encoding="utf-8") as f:
        for item in vaccine_list:
            sql = f"""INSERT INTO vaccine (
                vaccine_name,
                quantity,
                entry_date,
                expiry_date,
                manufacture_date,
                price
            ) VALUES (
                {sql_escape(item['vaccine_name'])},
                {int(item['quantity'])},
                {sql_date(item['entry_date'])},
                {sql_date(item['expiry_date'])},
                {sql_date(item['manufacture_date'])},
                {float(item['price'])}
            );\n"""
            f.write(sql)

    print(f"üíæ ƒê√£ l∆∞u {len(vaccine_list)} item v√†o {filename}")

vaccine_list = full_df.to_dict(orient="records")
write_sql(vaccine_list)
```
