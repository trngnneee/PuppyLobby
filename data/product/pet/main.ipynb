## Generate data pet cho tá»«ng Customer

```python
pet_species = ["dog", "cat", "bird", "fish", "hamster", "rabbit"]

breeds = {
    "dog": [
        "Labrador Retriever",
        "German Shepherd",
        "Golden Retriever",
        "Bulldog",
        "Beagle",
        "Poodle",
        "Rottweiler",
        "Yorkshire Terrier",
        "Boxer",
        "Dachshund"
    ],
    "cat": [
        "Siamese",
        "Persian",
        "Maine Coon",
        "Ragdoll",
        "Bengal",
        "Sphynx",
        "British Shorthair",
        "Abyssinian",
        "Scottish Fold",
        "Oriental Shorthair"
    ],
    "bird": [
        "Parakeet",
        "Cockatiel",
        "Lovebird",
        "Canary",
        "Finch",
        "African Grey",
        "Macaw",
        "Budgerigar",
        "Cockatoo",
        "Pionus"
    ],
    "fish": [
        "Goldfish",
        "Betta",
        "Guppy",
        "Angelfish",
        "Neon Tetra",
        "Molly",
        "Platy",
        "Swordtail",
        "Cichlid",
        "Zebrafish"
    ],
    "hamster": [
        "Syrian",
        "Roborovski",
        "Dwarf Campbell",
        "Winter White",
        "Chinese",
        "European",
        "Teddy Bear",
        "Fancy",
        "Short-haired",
        "Long-haired"
    ],
    "rabbit": [
        "Holland Lop",
        "Netherland Dwarf",
        "Flemish Giant",
        "Lionhead",
        "Mini Rex",
        "English Angora",
        "French Lop",
        "Dutch",
        "American Fuzzy Lop",
        "Hotot"
    ]
}

gender_choices = ["male", "female"]
health_status_choices = ["healthy", "sick", "recovering"]

pets = []
for cid in customer_ids:
    num_pets = random.randint(1, 3)
    for _ in range(num_pets):
        species = random.choice(pet_species)
        breed = random.choice(breeds[species])
        pet = {
            "pet_name": fake.first_name(),
            "species": species,
            "breed": breed,
            "age": random.randint(1, 15),
            "gender": random.choice(gender_choices),
            "health_state": random.choice(health_status_choices),
            "customer_id": cid
        }
        pets.append(pet)
```

```python
query = """
INSERT INTO pet (pet_name, species, breed, age, gender, health_state, customer_id)
VALUES %s
"""

values = [
    (a["pet_name"], a["species"], a["breed"], a["age"], a["gender"], a["health_state"], a["customer_id"])
    for a in pets
]

execute_values(cur, query, values, page_size=len(values))

conn.commit()
```
